```python
# Utilizando Pandas para abrir a planilha
import pandas as pd
import os

# Coletando o caminho da pasta que estou
cwd = os.getcwd()

# Pasta dos fatores de emiss√£o
path = cwd+'\\tabelas\\FE'
print(path)

# Indicando nome do arquivo
chapterNumer = 11
subChapter = 6
file = 'chp_'+str(chapterNumer)+'_'+str(subChapter)+'.csv'
print(file)

# Abrindo arquivo utilizando o Pandas
fe = pd.read_csv(path + '\\'+file)
print(fe)
```

```python
from datetime import datetime, timedelta

# Definindo a data inicial e final
data_inicial = datetime(2024, 1, 1, 0, 0)  # 1 de janeiro de 2024, 00:00
data_final = datetime(2024, 12, 31, 23, 0)  # 31 de dezembro de 2024, 23:00

# Definindo o passo (intervalo entre as datas)
passo = timedelta(hours=1)

# Criando o vetor de datas
vetor_datas = []
data_atual = data_inicial
while data_atual <= data_final:
    vetor_datas.append(data_atual)
    data_atual += passo

# Exibindo o vetor de datas
for data in vetor_datas:
    print(data)
```

```python
# Criando vetor de datas
import pandas as pd

# Criar um intervalo de datas
datas = pd.date_range(start='2024-01-01', end='2024-12-31', freq='h')  # Apenas um dia para exemplo

# Criar um DataFrame com as datas
df = pd.DataFrame({'Data': datas})

# Extrair os meses das datas e adicion√°-los como uma nova coluna no DataFrame
df['Mes'] = df['Data'].dt.month

# Extrair as horas das datas e adicion√°-las como uma nova coluna no DataFrame
df['Hora'] = df['Data'].dt.hour

# Mostrar o DataFrame resultante
print(df)
```

```python
# Calcular a soma das emiss√µes considerando todos os processos
labels = ['PM', 'SO2', 'NOX', 'CO', 'CO2', 'TOC']
for ii in range(1,len(fe.columns)):
    SomaFE = (fe.iloc[:,ii]).sum()
    print('Emiss√µes de ' + labels[ii-1])
    
    soma_emis = (df['aleatorioSazonal'].sum()*SomaFE*(1000/3600))
    # Calcular a emiss√£o m√©dia em g/s
    emis_med = df['aleatorioSazonal'].mean()*SomaFE*(1000/3600)
    # Calcular a emiss√£o m√≠nima em g/s
    emis_min = df['aleatorioSazonal'].min()*SomaFE*(1000/3600)
    # Calcular a emiss√£o m√°xima em g/s
    emis_max = df['aleatorioSazonal'].max()*SomaFE*(1000/3600)
    
    print("Somat√≥rio das emiss√µes: {:.2f} g/s".format(soma_emis))                                            
    print("Emiss√£o m√©dia: {:.2f} g/s".format(emis_med))
    print("Emiss√£o m√≠nima: {:.2f} g/s".format(emis_min))
    print("Emiss√£o m√°xima: {:.2f} g/s".format(emis_max))
    print(' ')
```
