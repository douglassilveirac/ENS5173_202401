# Numpy arrays

```python
import numpy as np

# Criando um vetor de dados aleat√≥rios
data = np.random.rand(1000,2)

# Verificando o tamanho do vetor
print(data.shape)
print(np.size(data))

# Print de uma amostra dos dados
print(data[0:10])
print(data[0:10].shape)
```

```python
# Usando Monte-Carlo
dataMax = 1000
dataMin = 10
data = (dataMax - dataMin)*data + dataMin
print(data[0:10])
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(3)
ax[0].plot(data)
ax[1].plot(data[:,0])
ax[2].plot(data[:,1])
print(type(data))
```

```python
# Criando um dado 2D 
data3d = np.sin(np.random.rand(1000,1000))
print(data3d)
```

```python
# Plotando um dado 3D
fig2, ax2 = plt.subplots()
ax2.pcolor(data3d)
```

```python
# Criando vertor temporal
dtime = np.arange(np.datetime64("2022-01-01"), np.datetime64("2023-01-01"), np.timedelta64(1, "h"))
print(dtime[1:10])
print(dtime.shape)
print(type(dtime))
```
