#### Calculando os parâmetros meteorológicos para o dia mais crítico do ano, onde todos os poluentes obtiveram a maior emissão (Dia 04/09/2023):

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from windrose import WindroseAxes

# Carregar o arquivo CSV com codificação 'latin1'
file_path = 'C:/Users/dodo_/Documents/GitHub/Atividade3/METEOcrit.csv'
data = pd.read_csv(file_path, encoding='latin1')

# Função para criar gráficos de linha para uma variável específica
def plot_hourly_data(data, column, ylabel):
    plt.figure(figsize=(12, 6))
    sns.lineplot(x='hora', y=column, data=data)
    plt.xlabel('Hora do Dia')
    plt.ylabel(ylabel)
    plt.xticks(range(0, 24))
    plt.grid(True)
    plt.show()

# Criar gráficos de linha para cada variável
plot_hourly_data(data, 'temperatura', 'Temperatura (K)')
plot_hourly_data(data, 'umidade', 'Umidade (%)')
plot_hourly_data(data, 'pressão', 'Pressão (hPa)')
plot_hourly_data(data, 'cobertura de nuves', 'Cobertura de Nuvens (oktas)')

# Função para criar a rosa dos ventos
def plot_wind_rose(data):
    plt.figure(figsize=(8, 8))
    ax = WindroseAxes.from_ax()
    ax.bar(data['direção do vento'], data['velocidade do vento'], normed=True, opening=0.8, edgecolor='white')
    ax.set_legend(title="Velocidade do Vento (m/s)")
    plt.show()

# Criar a rosa dos ventos
plot_wind_rose(data)
```
